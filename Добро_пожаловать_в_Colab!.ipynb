# Розгалуження, цикли, функції та виключення у Python3

Цей ноутбук містить нотатки та приклади коду для основних конструкцій Python3.


## 1. Розгалуження (if, elif, else)


```
x = 10

if x > 0:
    print("x додатнє")
elif x == 0:
    print("x дорівнює нулю")
else:
    print("x від'ємне")
```

## 2. Цикли (for та while)


```
# Цикл for
for i in range(1, 6):
    print('Цикл for:', i)

# Цикл while
count = 1
while count <= 5:
    print('Цикл while:', count)
    count += 1
```

## 3. Функції


```
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print('Факторіал 5:', factorial(5))
```

## 4. Виключення (try, except)


```
try:
    x = int(input('Введіть число: '))
    print('Ваше число:', x)
except ValueError:
    print('Помилка: введено не число!')
```

## 5. Інтегрований приклад: Меню для піднесення числа до квадрату


```
def square_number():
    try:
        num = int(input('Введіть число для піднесення до квадрату: '))
        print(f'{num}^2 = {num**2}')
    except ValueError:
        print('Помилка: введено не число!')

while True:
    print('\nМеню:')
    print('1 - Обчислити квадрат числа')
    print('2 - Вийти')
    choice = input('Виберіть опцію: ')

    if choice == '1':
        square_number()
    elif choice == '2':
        print('До побачення!')
        break
    else:
        print('Невірний вибір')
```
